```python
xs = [1, 2, 3]
ys = [1, 2, 2]
```

```python
import matplotlib.pyplot as plt

plt.scatter(xs, ys)
plt.xlabel("xs")
plt.ylabel("ys")
plt.title("Plot of xs vs ys")
plt.show()
```

```python
def f(x: float) -> float:
    return -(x**2) / 2 + 5 * x / 2 - 1
```

```python
import numpy as np

x_values = np.linspace(min(xs), 4, 100)
y_values = [f(x) for x in x_values]

plt.plot(x_values, y_values, label="f(x)")
plt.scatter(xs, ys, color="red")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of f(x)")
plt.legend()
plt.show()
```
